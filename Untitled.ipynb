```python
# Legal chatbot task - reviewing documents for legal compliance
task = '''
Please review the following document for legal compliance and provide feedback:
"Company X requires all employees to sign non-compete agreements that prevent them from working in the same industry for 5 years after termination. The company also reserves the right to monitor all employee communications without notice."
'''
```

```python
import autogen

# User Proxy Agent
user_proxy = autogen.UserProxyAgent(
    name="User_Proxy",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content", "").find("TERMINATE") >= 0,
    code_execution_config=False,
)
```

```python
# Specialized agents
legal_researcher = autogen.AssistantAgent(
    name="Legal_Researcher",
    system_message="You are a legal research assistant. Provide relevant laws, regulations, "
        "and case law references. Keep responses factual and reference reliable legal sources.",
    llm_config=llm_config,
)
```

```python
compliance_checker = autogen.AssistantAgent(
    name="Compliance_Checker",
    system_message="You specialize in regulatory compliance. Focus on GDPR, employment standards, "
        "and industry-specific regulations. Provide specific compliance recommendations.",
    llm_config=llm_config,
)
```

```python
client_communicator = autogen.AssistantAgent(
    name="Client_Communicator",
    system_message="You translate complex legal concepts into clear, understandable language for clients. "
        "Make legal advice accessible, actionable, and practical.",
    llm_config=llm_config,
)
```

```python
# Define the reflection_message function based on your original code logic
def reflection_message(recipient, messages, sender, config):
    """
    Reflection message function that follows the pattern from the original code.
    This function extracts the latest message content and asks the recipient to review it.
    """
    # Get the last message from the sender to the recipient
    last_message = recipient.chat_messages_for_summary(sender)[-1]['content']
    
    # Create a reflection prompt based on the recipient's role
    if recipient.name == "Legal_Researcher":
        return f'''As a legal researcher, review the following content and provide relevant legal research:
        
        {last_message}
        
        Please provide:
        1. Relevant laws and regulations
        2. Case law references if applicable
        3. Jurisdictional considerations
        4. Best practices in legal drafting'''
    
    elif recipient.name == "Compliance_Checker":
        return f'''As a compliance specialist, review the following content for regulatory compliance:
        
        {last_message}
        
        Please provide:
        1. Compliance issues identified
        2. Specific regulatory requirements
        3. Risk assessment
        4. Recommended compliance measures'''
    
    elif recipient.name == "Client_Communicator":
        return f'''As a client communication specialist, review the following legal content:
        
        {last_message}
        
        Please translate this into client-friendly advice that is:
        1. Accessible to non-lawyers
        2. Actionable and practical
        3. Prioritized by risk level
        4. Includes clear next steps'''
    
    else:
        # Default reflection message similar to your original code
        return f'''Review the following content and provide your specialized analysis: 
        
        {last_message}'''
```

```python
# Define the nested chats using the reflection_message function
review_chats = [
    {
        "recipient": legal_researcher, 
        "message": reflection_message, 
        "summary_method": "reflection_with_llm",
        "summary_args": {
            "summary_prompt": "Return your review as a JSON object: {'Reviewer': 'Legal Researcher', 'Review': 'Your review summary here'}"
        },
        "max_turns": 1
    },
    {
        "recipient": compliance_checker, 
        "message": reflection_message, 
        "summary_method": "reflection_with_llm",
        "summary_args": {
            "summary_prompt": "Return your review as a JSON object: {'Reviewer': 'Compliance Checker', 'Review': 'Your review summary here'}"
        },
        "max_turns": 1
    },
    {
        "recipient": client_communicator, 
        "message": reflection_message, 
        "summary_method": "reflection_with_llm",
        "summary_args": {
            "summary_prompt": "Return your review as a JSON object: {'Reviewer': 'Client Communicator', 'Review': 'Your review summary here'}"
        },
        "max_turns": 1
    }
]
```

```python
# Register the nested chats with the legal reviewer
legal_reviewer.register_nested_chats(
    review_chats,
    trigger=user_proxy,
)
```

```python
# Function to orchestrate the legal review process
def conduct_legal_review(document_text):
    print("=== STARTING LEGAL REVIEW ===")
    
    # Step 1: Initial legal review
    print("\n--- Legal Review ---")
    legal_review = user_proxy.initiate_chat(
        recipient=legal_reviewer,
        message=f"Please review this document for legal issues:\n\n{document_text}",
        max_turns=1,
        summary_method="last_msg",
        silent=True
    )
    
    # Step 2: Legal research
    print("\n--- Legal Research ---")
    legal_research = user_proxy.initiate_chat(
        recipient=legal_researcher,
        message=f"Based on this legal review: {legal_review.summary}\n\nPlease provide relevant legal research.",
        max_turns=1,
        summary_method="last_msg",
        silent=True
    )
    
    # Step 3: Compliance check
    print("\n--- Compliance Check ---")
    compliance_check = user_proxy.initiate_chat(
        recipient=compliance_checker,
        message=f"Based on this analysis: {legal_review.summary}\n\n{legal_research.summary}\n\nPlease conduct a compliance check.",
        max_turns=1,
        summary_method="last_msg",
        silent=True
    )
    
    # Step 4: Client communication
    print("\n--- Client Communication ---")
    client_advice = user_proxy.initiate_chat(
        recipient=client_communicator,
        message=f"Based on this complete analysis:\nLegal Review: {legal_review.summary}\nLegal Research: {legal_research.summary}\nCompliance Check: {compliance_check.summary}\n\nPlease provide client-friendly advice.",
        max_turns=1,
        summary_method="last_msg",
        silent=True
    )
    
    return {
        "legal_review": legal_review.summary,
        "legal_research": legal_research.summary,
        "compliance_check": compliance_check.summary,
        "client_advice": client_advice.summary
    }

# Execute the legal review
results = conduct_legal_review(task)

print("\n" + "="*60)
print("COMPREHENSIVE LEGAL REVIEW REPORT")
print("="*60)

# print("\n1. LEGAL REVIEW:")
# print(results["legal_review"])

# print("\n2. LEGAL RESEARCH:")
# print(results["legal_research"])

# print("\n3. COMPLIANCE CHECK:")
# print(results["compliance_check"])

# print("\n4. CLIENT ADVICE:")
# print(results["client_advice"])
```
